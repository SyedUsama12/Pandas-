```python
Customers= pd.DataFrame({"Customer NIC":[14,6,19,22],"customer name":["A","B","C","D"]})
Transactions= pd.DataFrame({"CNIC":np.random.choice([14,6,19,21],size=(20)),
                          "Transaction ID":np.random.randint(1,2,size=(20)),
                          "Transaction Date":pd.date_range("1990-01-01",periods=20,freq="M"),
                          "Description":np.random.choice(["TopUp","IBFT","FT"],size=(20)),
                          "Amount":np.random.randint(1,1000000,size=(20))})
print(Customers)
print(Transactions)
```

```python
c=Transactions.CNIC.value_counts()
#print(c.index)
#print(c.values)
display(c)
```

```python
a = Customers.merge(c,left_on="Customer NIC",right_index=True,how="left")
a.columns=["Customer cnic","customer name","Transaction counts"]
a.sort_values(by=["Transaction counts" ],ascending=False,inplace=True)
a
```

```python
b = Customers.merge(c,left_on="Customer NIC",right_index=True,how="right")
b.columns=["Customer cnic","customer name","Transaction counts"]
b.sort_values(by=["Transaction counts" ],ascending=False,inplace=True)
b
```

```python
f = Customers.merge(c,left_on="Customer NIC",right_index=True,how="inner")
f.columns=["Customer cnic","customer name","Transaction counts"]
f.sort_values(by=["Transaction counts" ],ascending=False,inplace=True)
f
```

```python
c = Customers.merge(c,left_on="Customer NIC",right_index=True,how="inner")
c.columns=["Customer cnic","customer name","Transaction counts"]
c.sort_values(by=["Transaction counts" ],ascending=False,inplace=True)
c
```

```python
e = Customers.merge(c,left_on="Customer NIC",right_index=True,how="outer")
e.columns=["Customer CNIC","Customer Name","Transaction counts"]
e.sort_values(by="Transaction counts" ,ascending=False)
```

# Stackplot

```python
days=[1,2,3,4,5]
reading=[3,6,8,9,9]
playing=[4,5,9,3,4]
watching=[13,7,6,8,12]
teaching=[2,15,20,5,4]
plt.plot([],[],label="days",color="c",linewidth=5)
plt.plot([],[],label="reading",color="m",linewidth=5)
plt.plot([],[],label="playing",color="r",linewidth=5)
plt.plot([],[],label="watching",color="k",linewidth=5)
plt.plot([],[],label="teaching",color="g",linewidth=5)
plt.stackplot(days,reading,playing,watching,teaching,colors=["c","m","r","k"])
plt.title("PIAIC data")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

# PieChart

```python
slices=[7,2,2,13]
activities=["Reading","Playing","Teaching","Watching"]
colors = ["r","c","y","k"]
plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=[0,0.1,0,0],autopct="%1.1f%%")
plt.title("PIAIC data")

plt.show()
```
