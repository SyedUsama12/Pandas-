```python
import numpy as np
import pandas as pd 
import pandas_datareader.data as web
```

```python
all_data = {Ticker:web.get_data_yahoo(Ticker) for Ticker in ["AAPL","IBM","MSFT","GOOG"] }
all_data
```

```python
all_data["AAPL"]
```
